# Feature Engineering

## qSOFA

```python
# Define a function to calculate the qSOFA score
def calculate_qsofa(row):
    score = 0
    score += (row['Resp'] >= 22)  # 1 point if respiratory rate >= 22/min
    score += (row['SBP'] <= 100)  # 1 point if systolic BP <= 100 mmHg
    # GCS is not provided, this condition is skipped
    return score
```

```python
# Apply the function to calculate qSOFA score for each row
df['Incomplete_qSOFA_Score'] = df.apply(calculate_qsofa, axis=1)
```

## SOFA SCORE

```python
class SOFAScoreCalculator:
    def __init__(self, row):
        self.row = row

    def respiratory_sofa(self, fio2, pao2):
        # Ensure FiO2 is not zero to avoid division by zero error
        fio2 = fio2 if fio2 != 0 else 1
        ratio = pao2 / fio2
        if ratio >= 400:
            return 0
        elif ratio < 400 and ratio >= 300:
            return 1
        elif ratio < 300 and ratio >= 200:
            return 2
        elif ratio < 200 and ratio >= 100:
            return 3
        else:
            return 4

    def cardiovascular_sofa(self, map):
        if map >= 70:
            return 0
        elif map < 70 and map >= 60:
            return 1
        elif map < 60 and map >= 50:
            return 2
        elif map < 50 and map >= 40:
            return 3
        else:
            return 4

    def liver_sofa(self, bilirubin_total):
        if bilirubin_total < 1.2:
            return 0
        elif bilirubin_total >= 1.2 and bilirubin_total < 2:
            return 1
        elif bilirubin_total >= 2 and bilirubin_total < 6:
            return 2
        elif bilirubin_total >= 6 and bilirubin_total < 12:
            return 3
        else:
            return 4

    def coagulation_sofa(self, platelets):
        if platelets >= 150:
            return 0
        elif platelets < 150 and platelets >= 100:
            return 1
        elif platelets < 100 and platelets >= 50:
            return 2
        elif platelets < 50 and platelets >= 20:
            return 3
        else:
            return 4

    def renal_sofa(self, creatinine):
        if creatinine < 1.2:
            return 0
        elif creatinine >= 1.2 and creatinine < 2:
            return 1
        elif creatinine >= 2 and creatinine < 3.5:
            return 2
        elif creatinine >= 3.5 and creatinine < 5:
            return 3
        else:
            return 4

    def calculate_sofa(self):
        resp_score = self.respiratory_sofa(self.row['FiO2'], self.row['SaO2'])
        cv_score = self.cardiovascular_sofa(self.row['MAP'])
        liver_score = self.liver_sofa(self.row['Bilirubin_total'])
        coagulation_score = self.coagulation_sofa(self.row['Platelets'])
        renal_score = self.renal_sofa(self.row['Creatinine'])
        return resp_score + cv_score + liver_score + coagulation_score + renal_score
```

```python
# Example of using the class
df['SOFA_Score'] = df.apply(lambda row: SOFAScoreCalculator(row).calculate_sofa(), axis=1)
```

## SIRS

```python
class SIRSCalculator:
    def __init__(self, row):
        self.row = row


    def sirs_criteria(self):
        criteria_met = 0
        # Body temperature criterion
        if self.row['Temp'] < 36 or self.row['Temp'] > 38:
            criteria_met += 1
        # Heart rate criterion
        if self.row['HR'] > 90:
            criteria_met += 1
        # Tachypnea criterion
        if self.row['Resp'] > 20 and (self.row['PaCO2'] <= 33 or ((self.row['MAP'] * self.row['FiO2']) / self.row['SaO2']) < 200):
            criteria_met += 1
        # WBC count criterion
        if self.row['WBC'] < 4 or self.row['WBC'] > 12:
            criteria_met += 1
        return 1 if criteria_met >= 2 else 0
```

```python
df['SIRS'] = df.apply(lambda row: SIRSCalculator(row).sirs_criteria(), axis=1)
```

## NEWS Score

```python
def calculate_NEWS(row):
    score = 0
    # Respiration Rate ('Resp')
    if row['Resp'] < 8 or row['Resp'] > 24:
        score += 3
    elif 8 <= row['Resp'] <= 11 or 21 <= row['Resp'] <= 24:
        score += 1

    # Oxygen Saturation ('O2Sat')
    if row['O2Sat'] < 90:  # Assuming need for supplemental oxygen if < 90%
        score += 3
    elif row['O2Sat'] < 92:
        score += 2

    # Systolic Blood Pressure ('SBP')
    if row['SBP'] < 90:  # Assuming MAP < 65 mmHg if SBP < 90 mmHg
        score += 3
    elif 90 <= row['SBP'] <= 100:
        score += 1

    # Pulse Rate ('HR')
    if row['HR'] < 40 or row['HR'] > 130:
        score += 3
    elif 41 <= row['HR'] <= 50 or 91 <= row['HR'] <= 110:
        score += 1

    # Temperature ('Temp')
    if row['Temp'] < 35 or row['Temp'] > 39:
        score += 1

    return score
```

```python
 #Apply the NEWS score calculation to each row
df['NEWS_Score'] = df.apply(calculate_NEWS, axis=1)
```

## APACHE II Score 

```python

def calculate_apache_ii_score(row):
    score = 0
    
    # Age scoring
    age = row['Age']
    if age <= 44:
        score += 0
    elif 45 <= age <= 54:
        score += 2
    elif 55 <= age <= 64:
        score += 3
    elif 65 <= age <= 74:
        score += 5
    else:
        score += 6

    # Temperature scoring
    temp = row['Temp']
    if temp >= 41:
        score += 4
    elif 39 <= temp < 41:
        score += 3
    elif 38.5 <= temp < 39:
        score += 1
    elif 36 <= temp < 38.5:
        score += 0
    elif 34 <= temp < 36:
        score += 1
    elif 32 <= temp < 34:
        score += 2
    elif 30 <= temp < 32:
        score += 3
    else:
        score += 4
    
    # MAP scoring
    map = row['MAP']
    if map > 159:
        score += 4
    elif 130 <= map <= 159:
        score += 3
    elif 110 <= map <= 129:
        score += 2
    elif 70 <= map <= 109:
        score += 0
    elif 50 <= map <= 69:
        score += 2
    else:
        score += 4
    
    # Heart rate scoring
    hr = row['HR']
    if hr >= 180:
        score += 4
    elif 140 <= hr < 180:
        score += 3
    elif 110 <= hr < 140:
        score += 2
    elif 70 <= hr < 110:
        score += 0
    elif 55 <= hr < 70:
        score += 2
    elif 40 <= hr < 55:
        score += 3
    else:
        score += 4
    
    # Respiratory rate scoring
    resp = row['Resp']
    if resp >= 50:
        score += 4
    elif 35 <= resp < 50:
        score += 3
    elif 25 <= resp < 35:
        score += 1
    elif 12 <= resp < 25:
        score += 0
    elif 10 <= resp < 12:
        score += 1
    elif 6 <= resp < 10:
        score += 2
    else:
        score += 4

    # Oxygenation scoring - requires FiO2 and either PaO2 or SaO2
    # This part is a bit tricky because the scoring depends on the PaO2/FiO2 ratio or SaO2
    # For simplicity, I'm assuming will use PaO2 and FiO2 to calculate the ratio
    fio2 = row['FiO2'] if row['FiO2'] != 0 else 1  # Prevent division by zero
    pao2 = row['SaO2']  
    ratio = pao2 / fio2
    if ratio >= 400:
        score += 0
    elif 300 <= ratio < 400:
        score += 1
    elif 200 <= ratio < 300:
        score += 2
    elif 100 <= ratio < 200:
        score += 3
    else:
        score += 4
    
    # Arterial pH scoring
    ph = row['pH']
    if ph >= 7.7:
        score += 4
    elif 7.6 <= ph < 7.7:
        score += 3
    elif 7.5 <= ph < 7.6:
        score += 1
    elif 7.33 <= ph < 7.5:
        score += 0
    elif 7.25 <= ph < 7.33:
        score += 2
    elif 7.15 <= ph < 7.25:
        score += 3
    else:
        score += 4
    
  
    
    # Serum potassium scoring
    potassium = row['Potassium']
    if potassium >= 7.0:
        score += 4
    elif 6.0 <= potassium < 7.0:
        score += 3
    elif 5.5 <= potassium < 6.0:
        score += 1
    elif 3.5 <= potassium < 5.5:
        score += 0
    elif 3.0 <= potassium < 3.5:
        score += 1
    elif 2.5 <= potassium < 3.0:
        score += 2
    else:
        score += 4
    
    # Serum creatinine scoring
    creatinine = row['Creatinine']
    if creatinine >= 3.5:
        score += 4
    elif 2.0 <= creatinine < 3.5:
        score += 2  # Note: the criteria might have been duplicated; adjusting based on standard Apache II scoring
    elif 1.5 <= creatinine < 2.0:
        score += 1
    elif 0.6 <= creatinine < 1.5:
        score += 0
    else:
        score += 2
    
    # Hematocrit scoring
    hct = row['Hct']
    if hct >= 60:
        score += 4
    elif 50 <= hct < 60:
        score += 2
    elif 46 <= hct < 50:
        score += 1
    elif 30 <= hct < 46:
        score += 0
    elif 20 <= hct < 30:
        score += 2
    else:
        score += 4
    
    # White blood cell count scoring
    wbc = row['WBC']
    if wbc >= 40:
        score += 4
    elif 20 <= wbc < 40:
        score += 2
    elif 15 <= wbc < 20:
        score += 1
    elif 3 <= wbc < 15:
        score += 0
    elif 1 <= wbc < 3:
        score += 2
    else:
        score += 4
    
    return score

```

```python
# To apply this function and create a new column 'Apache_II_Score' in the DataFrame:
df['Apache_II_Score'] = df.apply(calculate_apache_ii_score, axis=1)
```

## Calculating BUN/creatine ration

```python

# Calculate BUN/Creatinine ratio and add as a new column
df['BUN_Creatinine_Ratio'] = df['BUN'] / df['Creatinine']
```

## Shcock index and Bilirubin creatinine ratio 

```python
# Calculate the Shock Index: Heart Rate / Systolic Blood Pressure
df['Shock_Index'] = df['HR'] / df['SBP']

# Calculate the ratio of Bilirubin to Creatinine
df['Bilirubin_Creatinine_Ratio'] = df['Bilirubin_total'] / df['Creatinine']
```

## Define start time

```python

# Define the start time
start_time = pd.Timestamp('2019-01-01 01:00:00')

# Convert 'hour' to datetime for the first 1000 rows
for patient_id, group in df.head(1000).groupby('Patient_ID'):
    # Calculate the datetime based on the 'hour' value
    group['datetime'] = start_time + pd.to_timedelta(group['Hour'], unit='h')
    
    # Update the DataFrame with the converted datetime
    df.loc[group.index, 'datetime'] = group['datetime']
```
